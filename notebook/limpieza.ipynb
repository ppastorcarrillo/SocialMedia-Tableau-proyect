```python
a = pd.read_csv("social media influencers - instagram sep-2022.csv", index_col=0)
```

```python
def limpiar_num(col):
    if "K" in col:
        col = float(col.replace("K",""))
        return col/1000
    elif "M" in col:
        return float(col.replace("M",""))  
```

```python
columna = a["Engagement average\r\n"].apply(limpiar_num)
```

```python
columna
```

```python
a["Engagement average\r\n(M)"] = columna
```
